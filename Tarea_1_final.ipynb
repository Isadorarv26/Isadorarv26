---


El Producto Interno Bruto (GDP) per cápita de un país es un indicador que representa la productividad económica por habitante. Este ratio se obtiene dividiendo el GDP por el total de población de determinado país. Uno de los métodos para cálculo del GDP es a través de la demanda agregada de bienes en la economía, de esta manera es posible obtener el ratio por habitante. Las fórmulas correspondientes son las siguientes:

$$ GDP = C + I + G + (X - M) $$

$$ GDP\ \text{(per capita)} = \frac{GDP}{\text{Habitantes}} $$


`Population`indica el número de habitantes por país. Es probable que ante mayor población se genere mayor actividad/producción económica, lo cual estimula la demanda agregada de bienes e incrementa el nivel de GDP, pero además aumenta el denominador del indicador per cápita. Si bien por esto último, mayor población debiera relacionarse con mayor producto per cápita, la evidencia demuestra que dicha variable tiene mayor relación con factores que afecten la de productividad o el bienestar de un país.

Un país con mayor `Area (sq. mi.)` tendrá mayor extensión de territorio, ya sea destinado a áreas comerciales o extracción/explotación de materias primas, por ende tendrá una producción interna mayor, y así un mayor GDP per cápita.

La variable `Pop. Density (per sq. mi.)` indica densidad de población por milla. Graficos de correlación (OWD, 2023) contra `GDP per cápita` demuestran que no hay una relacion significativa entre las dos variables.

`Net migration` corresponde a la resta entre el número de inmigrantes y emigrantres, para un país dentro de un período de tiempo. El indicador se encuentra normalizado como proporción de la población total por cada 1000 habitantes (INE, s.f.). Una mayor migración neta se entiende como mayor entrada de trabajadores al país respecto a los que salieron, por lo que mayor fuerza laboral implica a priori mayor producción económica, y por ende mayor GDP per cápita.

La variable `Infant mortality (per 1000 births)`presenta un índice de cantidad de mortalidad infantil normalizado a 1000 nacimientos. Mayores números en esta variable puede indicar países menos desarrollados en la industria de salud y económicamente, por lo que se puede relacionar con menores niveles de GDP per cápita.

`Literacy (%)` como alfabetización, representa el porcentaje de la población con capacidad de leer y escribir. Esto puede ser asociado como un índice de capital humano fundamental para el desarrollo de los países, donde países más desarrollados poseen una tasa de alfabetización alta. Es por esto que la variable tiene correlación positiva con el GDP per cápita.

`Phones (per 1000)` representa la cantidad de teléfonos por cada 1000 personas. El acceso a dispositivos móviles puede indicar el nivel de conectividad y acceso a la tecnología del país, lo cual podría fomentar el desarrollo y producción de los países. La variable puede estar asociada positivamente con el GDP per cápita.

Se presentan variables porcentuales relacionadas a la actividad agrícola de cada país, tales como porcentaje de tierra arable `Arable (%)`y porcentaje de tierra cultivable `cultivable (%)`. Evidencia (OWD, 2023) demuestra que países donde el setor agrícola representa una mayor proporción del GDP suelen tener menores niveles de GDP per cápita. Al contario, la variable `Other (%)` representa porcentualmente el uso alternativo del suelo respecto la agricultura (supuesto), por lo que podría estar relacionado positivamente con GDP per cápita.


Hay variables demográficas de cada país, tasa de natalidad `Birthrate` y tasa de mortalidad `Birthrate`. La tendencia global de los países desarrollados es una baja en la tasa de natalidad (World Bank, 2022), por lo que esta variable puede ser relacionada negativamente con GDP per cápita. Una alta tasa de mortalidad puede asociarse con sistemas de salud precarios y menor calidad de vida, características comúnmente asociadas a países con menor desarrollo y por ende con menor GDP per cápita.

Si en 2007 no se reporta una fuente de emisión para un país, por ejemplo, si emisiones de `Gas Flaring` no se reportara, entonces asigne números de 1 a 4 a las fuentes restantes. Análogo para un menor número de datos.

Luego, por cada variable grafique un histograma de frecuencias del ranking que obtuvo la fuente emisión a lo largo de todos los países.

¿Cuál fue la fuente más contaminante en la mayoría de países en 2007?


```python
df_co2_2007 = df_co2[df_co2['Year'].dt.year == 2007]

df_co2_2007 = df_co2_2007.drop(columns=['Per Capita', 'Bunker fuels (Not in Total)'])

sources = ['Solid Fuel', 'Liquid Fuel', 'Gas Fuel', 'Cement', 'Gas Flaring']

print(df_co2_2007)
```

```python
rankings = []

#ranking de fuentes de emisión por país
for country in df_co2_2007['Country'].unique():
    country_data = df_co2_2007[df_co2_2007['Country'] == country]
    present_sources = [source for source in sources if source in country_data.columns and not pd.isnull(country_data[source].values[0])]

    if len(present_sources) > 0:

        sorted_sources = country_data[present_sources].iloc[0].sort_values(ascending=False)
        ranking = {source: rank + 1 for rank, source in enumerate(sorted_sources.index)}
        rankings.append(ranking)

#df rankings
rankings_df = pd.DataFrame(rankings, index=df_co2_2007['Country'].unique())

rankings_df.info()
rankings_df.head()
rankings_df.tail()
```

```python
if not rankings_df.empty:
    #fuente más contaminante 2007
    most_common_rank_1 = rankings_df.eq(1).sum().idxmax()
    print(f"La fuente más contaminante en la mayoría de los países en 2007 fue: {most_common_rank_1}")
else:
    print("No hay datos suficientes para calcular el ranking.")

```

```python
#histogramas
plt.figure(figsize=(12, 10))
for idx, source in enumerate(sources):
    if source in rankings_df.columns:
        plt.subplot(3, 2, idx + 1)
        plt.hist(rankings_df[source].dropna(), bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5], alpha=0.7, edgecolor='black')

        plt.title(f'Histograma de Rankings para {source}')
        plt.xlabel('Ranking de Emisión (1 = Mayor Emisión)')
        plt.ylabel('Número de Países')

        #enteros
        plt.xticks([1, 2, 3, 4, 5])

plt.tight_layout()
plt.show()
```
